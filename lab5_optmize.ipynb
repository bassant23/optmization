```python
def Gradient(x0, max_it,alpha , df ):
    x=x0
     
    for i in range(max_it):
        print(f"\n******* Iteration {i} *******")
        x = x -  (df(x) *alpha)
        print("x_min = ",x)
        grad = np.abs(df(x))
        print("Gradient = " ,grad)
        if grad < 0.1:
            break
        
     
    
    return x , grad ,i
```

```python
x_min , grad , i = Gradient(40, 400, 0.01 ,Df)
```

```python
# Gradient descent with learning rate = 0.01, x0=40, epsilon = 0.1
```

```python
def NewtonMethod(x0, max_it, f ,df):
    x=x0
     
    for i in range(max_it):
        print(f"\n*** Iteration {i} ***")
        x=x-f(x)/df(x)
        print("x_min = ",x)
        residual=np.abs(f(x))
        print("Gradient = " ,residual)
        if residual < 0.01:
            break
        
     
    
    return x,residual
```

```python
f_1 = lambda x: x**2  + y**2
Df_1 = lambda X: np.array([2*X[0],2*X[1]])
DDf_1 = lambda X: np.array([[2,0],
                         [0,2]])
```

```python
f_2 = lambda x: x**2 + x*y + y**2
Df_2 = lambda X: np.array([2*X[0]+X[1],2*X[1]+X[0]])
DDf_2 = lambda X: np.array([[2,1],
                        [1,2]])
```

```python
def Gradient_Multi(x0, max_it,alpha ,epsilon, df ):
    x=x0
     
    for i in range(max_it):
        print(f"\n******* Iteration {i} *******")
        x = x -  (df(x) *alpha)
        grad = np.abs(df(x))
        if grad[0] < epsilon and grad[1] < epsilon:
            break
        
    print(f"\n\n******* Training Report *******")  
    print(f"Found solution after {i+1} iterations.")
    print("xy_min =" , x)
    print('Gradient=', grad )
     
    
    return x , grad ,i
```

```python
x_min , grad , i = Gradient_Multi( (10,10) , 400,  0.1, 0.001 , Df_1)
```

```python
def Newton(x0, max_it ,alpha, epsilon, df ,ddf):
    x=x0
     
    for i in range(max_it):
        x = x- alpha*(np.linalg.inv(ddf(x)) * df(x))
        grad = np.abs(df(x))
        if grad[0,0] < epsilon and grad[1,1] < epsilon:
            break
        
    print(f"\n\n******* Training Report *******")  
    print(f"Found solution after {i+1} iterations.")
    print("xy_min =" , np.diag(x))
    print('Gradient=',np.diag(grad) )
    
    return x , grad ,i
 
```

```python
x_min , gred ,i = Newton((10,10), 200 ,1, 0.001 , Df_1 , DDf_1)
```
